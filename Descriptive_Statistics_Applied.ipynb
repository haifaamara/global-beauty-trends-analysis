# **Visualisation:**

```python
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_distribution(df, numerical_columns):
    """
    Visualizes the distribution of numerical columns in a dataset using Histogram, Box Plot, and KDE Plot.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the data.
        numerical_columns (list): List of numerical column names to visualize.

    Returns:
        None
    """
    for column in numerical_columns:
        # Create a figure for each column
        fig, axes = plt.subplots(1, 3, figsize=(12, 2))

        # Histogram
        sns.histplot(df[column], kde=False, bins=10, color='blue', alpha=0.7, ax=axes[0])
        axes[0].set_title(f"Histogram of {column}", fontsize=14)
        axes[0].set_xlabel(column, fontsize=12)
        axes[0].set_ylabel("Frequency", fontsize=12)

        # Box Plot
        sns.boxplot(x=df[column], color='yellow', ax=axes[1])
        axes[1].set_title(f"Box Plot of {column}", fontsize=14)
        axes[1].set_xlabel(column, fontsize=12)

        # KDE Plot
        sns.kdeplot(df[column], fill=True, color='red', alpha=0.6, ax=axes[2])
        axes[2].set_title(f"KDE Plot of {column}", fontsize=14)
        axes[2].set_xlabel(column, fontsize=12)
        axes[2].set_ylabel("Density", fontsize=12)

        # Adjust layout and display the plots
        plt.tight_layout()
        plt.show()

# Example Usage:
Makeup_data = pd.read_csv('most_used_products.csv')
# Specify numerical columns
numerical_columns = ['Price_USD', 'Rating', 'Number_of_Reviews', 'Product_Size_ml']

# Call the function to visualize the distribution of the specified columns
visualize_distribution(Makeup_data, numerical_columns)
```

# **Distribution; Skewness & Kurtosis:**

```python
def calculate_distribution_metrics(df, numerical_columns):

    distribution_metrics = {}
    for column in numerical_columns:
        distribution_metrics[column] = {
            'Skewness': df[column].skew(),
            'Kurtosis': df[column].kurtosis()
        }
    return pd.DataFrame(distribution_metrics)
```

```python
# Numerical columns for analysis
numerical_columns = ['Price_USD', 'Rating', 'Number_of_Reviews', 'Product_Size_ml']

# Calculate distribution metrics
distribution_df = calculate_distribution_metrics(Makeup_data, numerical_columns)

print(distribution_df)
```

# **Dispersion Analysis:**

* **Range:** Indicates the total spread of the data.
* **Variance:** Quantifies the overall variability in the dataset.
* **Standard Deviation (SD):** Provides variability in the same units as the original data.
* **Interquartile Range (IQR)**: Captures the spread of the middle 50% of the data, robust to outliers.



```python
def calculate_dispersion(df, numerical_columns):
    dispersion_measures = {}

    for column in numerical_columns:
        dispersion_measures[column] = {
            'Range': df[column].max() - df[column].min(),
            'Variance': df[column].var(),
            'Standard Deviation': df[column].std(),
            'IQR': df[column].quantile(0.75) - df[column].quantile(0.25)
        }

    return pd.DataFrame(dispersion_measures).T

# List of numerical columns to analyze
numerical_columns = ['Price_USD', 'Rating', 'Number_of_Reviews', 'Product_Size_ml']
# Calculate dispersion measures
dispersion_df = calculate_dispersion(Makeup_data, numerical_columns)

dispersion_df
```
